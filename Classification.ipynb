The data for this project is in a csv format. The following describes the columns present in the data.

## Hypothesis:

### *General Hypothesis*
Null Hypothesis (H0): There is no significant difference in the behavior and characteristics of customers who churn and customers who do not churn.

Alternative Hypothesis (H1): There are significant differences in the behavior and characteristics of customers who churn and customers who do not churn.


### *Specific Hypothesis*
**Hypothesis 1:**

Null Hypothesis (H0): Customer tenure (length of subscription) has no significant effect on the likelihood of churn.

Alternative Hypothesis (HA): Customer tenure is significantly associated with the likelihood of churn, with longer tenured customers being less likely to churn.

**Hypothesis 2:**

Null Hypothesis (H0): There is no significant difference in the churn rates between customers with and without additional services (OnlineSecurity, TechSupport, DeviceProtection, StreamingTV, StreamingMovies).

Alternative Hypothesis (HA): Customers who have additional services are less likely to churn compared to customers without these services.

**Hypothesis 3:**

Null Hypothesis (H0): Gender has no impact on customer churn rates.

Alternative Hypothesis (HA): There is a significant difference in churn rates between male and female customers, with one gender being more likely to churn than the other.

How does the distribution of churn ('Yes' and 'No', 0 and 1) vary across different categories of the predictor variable?

```python
# Convert "Churn" column to numerical values (1 for "Yes" and 0 for "No")
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Define a function to calculate churn rate for each category in a column
def calculate_churn_rate(column):
    churn_rates = df.groupby(column)['Churn'].mean() * 100
    return churn_rates


for column in df.drop(columns=['Churn', 'MonthlyCharges', 'TotalCharges']).columns:
    churn_rates = calculate_churn_rate(column)
    print(f"\nHow does the churn rate vary in the {column} column?")
    for category, rate in churn_rates.items():
        print(f"{category}: Churn Rate = {rate:.2f} %")
    print(f"Total Customers in each category:\n{df[column].value_counts()}\n")

    # Create a bar chart to visualize churn rates for each category in the column
    plt.figure(figsize=(12, 6))
    sns.barplot(x=churn_rates.index, y=churn_rates.values, palette='coolwarm')
    plt.xlabel(column)
    plt.ylabel('Churn Rate (%)')
    plt.title(f'Churn Rate Variation by {column}')
    plt.show()
```

We can learn a lot from these charts. Here are some insights:

1. Customers without dependents are two times more likely to churn.

2. Customers that use paperless billing and optical fiber are more likely to churn.

3. Customers with no online security or backup, no device protection, and no tech support are from two to three times more likely to churn.

4. Customers with no internet service are unlikely to churn.

5. Customers with month-to-month contracts are almost four times more likely to churn than customers with yearly contracts. Two-year contractors are very unlikely to churn.

6. Customers that use electronic checks to pay their bills are more likely to churn.

7. We can see there is a significant correlation between the tenure and Churn columns. The highest the tenure, the lowest the chances that the customer will churn.

```python
# Count the occurrences of 'Yes' and 'No' in the churn data
churn_counts = testdataset['Churn_Prediction'].value_counts()
# Data to plot
labels = churn_counts.index
sizes = churn_counts.values
colors = ['green', 'darkorange']
explode = (0.1, 0)  # To explode the first slice (Churned)

# Plot the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

# Add a title
plt.title('Churn_Prediction Data Pie Chart')

# Display the chart
plt.show()
```

How does the distribution of churn ('Yes' and 'No', 0 and 1) vary across different categories of the predictor variable?

```python
# Convert "Churn" column to numerical values (1 for "Yes" and 0 for "No")
testdataset['Churn_Prediction'] = testdataset['Churn_Prediction'].map({'Yes': 1, 'No': 0})

# Define a function to calculate churn rate for each category in a column
def calculate_churn_rate(column):
    churn_rates = testdataset.groupby(column)['Churn_Prediction'].mean() * 100
    return churn_rates


for column in testdataset.drop(columns=['Churn_Prediction', 'MonthlyCharges', 'TotalCharges']).columns:
    churn_rates = calculate_churn_rate(column)
    print(f"\nHow does the churn_prediction rate vary in the {column} column?")
    for category, rate in churn_rates.items():
        print(f"{category}: Churn_Prediction Rate = {rate:.2f} %")
    print(f"Total Customers in each category:\n{testdataset[column].value_counts()}\n")

    # Create a bar chart to visualize churn rates for each category in the column
    plt.figure(figsize=(12, 6))
    sns.barplot(x=churn_rates.index, y=churn_rates.values, palette='coolwarm')
    plt.xlabel(column)
    plt.ylabel('Churn_Prediction Rate (%)')
    plt.title(f'Churn_Prediction Rate Variation by {column}')
    plt.show()
    # print(......................................................................................................)
```
